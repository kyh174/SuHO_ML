```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np

plt.rc('font', family='Malgun Gothic')

df = pd.read_csv('./Data/인구예측 찐.csv')

# plt.figure(figsize=(6,6))
# plt.scatter(df['자연증가건수'], df['인천인구수'])

x_data = np.array(df['자연증가건수']).reshape(-1, 1)
y_data = np.array(df['인천인구수'])

# x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(x_data, y_data)
print(model.score(x_data, y_data))
# print(model.score(x_train, y_train))

coef = model.coef_
intercept = model.intercept_

pred = coef * x_data + intercept

# print(y_test)
# print(model.predict(x_test))

plt.figure(figsize=(6,6))
plt.scatter(x_data, y_data)
plt.plot(x_data, model.predict(x_data), c='r')

# df
```

```python
plt.figure(figsize=(10,10))
plt.plot(df['연도'], df['출생건수'], label='출생')
plt.plot(df['연도'], df['사망건수'], label='사망')
plt.plot(df['연도'], df['자연증가건수'], label='자연증가')
plt.legend()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=6, include_bias=False)
x_poly = poly_features.fit_transform(x_data)

# print(x_data[0])
# print(x_poly[0])

lin_model = LinearRegression()
lin_model.fit(x_poly, y_data)
d_coef = lin_model.coef_
d_intercept = lin_model.intercept_

point = np.linspace(0,35000,29).reshape(29,1)
p_poly = poly_features.fit_transform(point)

plt.figure(figsize=(6,6))
plt.scatter(x_data, y_data)
plt.plot(point, lin_model.predict(p_poly), c='r')

print(lin_model.score(x_poly, y_data))



# print('ssss', int(lin_model.predict([[494516**6 ,494516**5, 494516**4, 494516**3, 494516**2, 494516]])))
```
