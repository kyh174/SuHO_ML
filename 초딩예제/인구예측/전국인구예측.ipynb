```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np

plt.rc('font', family='Malgun Gothic')

df = pd.read_csv('./Data/전국_인구예측(출생,사망_인구).csv')

# 모델 학습위해 전처리
x_data = np.array(df['자연증가건수']).reshape(-1, 1)
y_data = np.array(df['총인구수'])
# 모델 생성
model = LinearRegression()
model.fit(x_data, y_data)
print('단순회귀 평가점수: ', model.score(x_data, y_data))

# 예측선 그리기
coef = model.coef_
intercept = model.intercept_
pred = coef * x_data + intercept
# 예측선 시각화
plt.figure(figsize=(6,6))
plt.scatter(x_data, y_data)
plt.plot(x_data, model.predict(x_data), c='r')
```

```python
# 다항회귀
from sklearn.preprocessing import PolynomialFeatures

# 다항모델 생성 및 변환
poly_features = PolynomialFeatures(degree=6, include_bias=False)
x_poly = poly_features.fit_transform(x_data)

# 선형회귀모델
model2 = LinearRegression()
model2.fit(x_poly, y_data)
p_coef = model2.coef_
p_intercept = model2.intercept_
print('다항회귀 평가점수: ', model2.score(x_poly, y_data))

plt.figure(figsize=(6,6))
plt.scatter(x_data, y_data)
plt.plot(x_data, model2.predict(x_poly), c='r')
```
