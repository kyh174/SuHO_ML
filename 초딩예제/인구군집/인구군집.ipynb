```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
plt.rc('font', family='Malgun Gothic')

df = pd.read_csv('./Data/인천시_구별_주민등록세대수.csv', index_col='행정구역(시군구)별')

# df.isna().sum()
df = df.drop(index=['인천광역시', '미추홀구', '북구'], axis=0)
# # df.info()
# df.isna().sum()
df = df.fillna(df.mean())
# df.isna().sum()



# 연도별 인천시 시각화
# plt.figure(figsize=(10,10))
# plt.plot(df['시점'], df['인천광역시'])

# 연도별 구 시각화
# 이건 행렬변환을 해서 그래프를 그려야될거같은데
col_year = df.columns
# plt.figure(figsize=(10,10))
# for year in col_year:
#     print(year)
#     plt.bar(df['행정구역(시군구)별'], df[year])
# plt.legend()


# 군집
model = KMeans(n_clusters=3)
model.fit(df[col_year])
df['cluster'] = model.labels_



# df[df['cluster'] == 0]

df
```
